```python
#Handle Nullable Values
df = df.dropna(subset = ['continent','location' ])
df['population']=df['population'].fillna(df['population'].mode()[0])
df['total_deaths']=df['total_deaths'].fillna(df['total_deaths'].mode()[0])
df['total_cases']=df['total_cases'].fillna(df['total_cases'].mode()[0])
df['total_vaccinations']=df['total_vaccinations'].fillna(df['total_vaccinations'].mode()[0])

df['human_development_index']=df['human_development_index'].fillna(df['human_development_index'].mode())
df['icu_patients']=df['icu_patients'].fillna(df['icu_patients'].mode())
df['stringency_index']=df['stringency_index'].fillna(df['stringency_index'].mode())

# df['reproduction_rate']=df['reproduction_rate'].fillna(df['reproduction_rate'].median())
df['total_tests']=df['total_tests'].fillna(df['total_tests'].mode())
df['handwashing_facilities']=df['handwashing_facilities'].fillna(df['handwashing_facilities'].mode())

df['median_age']=df['median_age'].fillna(df['median_age'].median())
df['aged_70_older']=df['aged_70_older'].fillna(df['aged_70_older'].median())
df['female_smokers']=df['female_smokers'].fillna(df['female_smokers'].median())
df['male_smokers']=df['male_smokers'].fillna(df['male_smokers'].median())

```

```python
df = df[(df['date'] == '2021-10-26')]
# df.head()
```
